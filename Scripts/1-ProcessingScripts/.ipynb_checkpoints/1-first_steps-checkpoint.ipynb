## 3-Histórico Asesor

Se trata de una base con 55482 filas y 9 variables. El ella hay 25891 personas únicas. Debemos calcular el número de veces que la persona ha cambiado de asesor, o lo que es lo mismo, crear una nueva variable `n_asesores_hist` que cuente cuantos asesores ha tenido cada persona. 

```python
print(historico_asesor.shape)
print(historico_asesor.columns)
```

```python
historico_asesor['afi_hash64'].value_counts()
```

### Anomalías
#### Primera
Encuentro que en ciertos individuos (identificados con el `afi_hash64`) que aparecen varias veces en la base (e.g. individuo "763c92a034fe8b83c227fc5ca330c8fd62a07d25" que aparece 131 veces), lo que debería indicar que han hecho varios cambios de asesor, en cada registro de cambio aparecen `SIN ASIGNACIÓN` en la variable sobre el cargo del asesor al que se cambian. Esto significa que a pesar de no tener asesor (de no tenerlo en ninguno de los registros), se registró su "cambio de asesor" o su "histórico" de asesor" en la base.

#### Segunda
Hay cuentas que fueron creadas en el 2017 y hasta 2018. Esto es extraño dado que en la base de cancelados y de suscritos las fechas más tempranas son del 2019 y 2020 respectivamente.

```python
historico_asesor['eliminar'].value_counts()
```

```python
print(historico_asesor['cargo de asesor al que se cambia'].value_counts())
historico_asesor['cargo de asesor al que se cambia'].value_counts(normalize=True).plot.pie(autopct='%1.1f%%')
plt.ylabel('')
plt.title('Asesor al que se cambia (histórico asesor)')
```

#### Revisión primera anomalía
Sospecho que tal vez hay muchos registros (filas) en la base de histórico asesor que en realidad no son un *cambio* de asesor, sino simplemente registros de la misma persona, tal vez con diferentes cuentas, **pero con el mismo 'cargo de asesor al que se cambia'**, es decir `SIN ASIGNACION`, que representa el 90.3% de las observaciones en esta variable.

```python
filtro1 = ~(historico_asesor[['afi_hash64','cargo de asesor al que se cambia' ]].duplicated(keep='first')) #mark as True all non-duplicates 
df_step1 = historico_asesor[filtro1]
filtro2 = ~(df_step1['afi_hash64'].duplicated(keep=False)) #mark as True all unique observations
df_step2= df_step1[filtro2]['cargo de asesor al que se cambia']
print('Hay', df_step2.shape[0], 'observaciones que no cambiaron de asesor o estado')
print(df_step2.value_counts(normalize=True))
print('\nDe estas',df_step2.shape[0], 'el 89% estuvieron sin asesor')
```

Por lo descrito anteriormente se considera inadecuado tener un conteo de las veces que la persona ha cambiado de asesor, pues habrá un sobre-conteo de aquellos registros que son solo `SIN ASIGNACION` pero registrados varias veces.

### Varianza 

Se requiere construir una variable numérica categórica para poder calcular la varianza de perfiles de riesgo de las asignaciones de cada sujeto. Para ello se seguirá la siguiente regla:

1. 1 = 'Muy Conservador' o 'Cauteloso'
2. 2 = 'Moderado'
3. 3 = 'Valiente' o 'Arriesgado'
4. 4 = 'Capacidad' 1 a 5
5. 5 = 'Conocimiento' 1 a 5
6. 6 = 'Tolerancia' 1 a 5

A esta variable nueva `riesgo_categ` se le calculará el logaritmo para evitar que la diferencia entre el número más bajo (1) y el más alto (6) sea tan significativa al calcular la varianza. Esto porque **la escala no es ordenada**, i.e. un nivel 6 (Tolerancia 1 a 5) no implica seis veces (x6) el nivel 1 (Muy Consdervador).

```python

condiciones = [(perfil_riesgo_hist['Perfil de riesgo resultante de actualización'] == 'Muy Conservador') | (perfil_riesgo_hist['Perfil de riesgo resultante de actualización'] == 'Cauteloso'),
               perfil_riesgo_hist['Perfil de riesgo resultante de actualización'] == 'Moderado',
               (perfil_riesgo_hist['Perfil de riesgo resultante de actualización'] == 'Valiente') | (perfil_riesgo_hist['Perfil de riesgo resultante de actualización']=='Arriesgado'),
               perfil_riesgo_hist['Perfil de riesgo resultante de actualización'].str.contains('Capacidad'),
               perfil_riesgo_hist['Perfil de riesgo resultante de actualización'].str.contains('Conocimiento'),
               perfil_riesgo_hist['Perfil de riesgo resultante de actualización'].str.contains('Tolerancia')]
values = [1, 2, 3, 4, 5, 6]
perfil_riesgo_hist['riesgo_categ'] = np.select(condiciones, values)
perfil_riesgo_hist['riesgo_categ_log']= np.log(perfil_riesgo_hist['riesgo_categ'])
perfil_riesgo_hist['riesgo_categ'].value_counts()
```

```python
# Creo variable para guardar varianza
perfil_riesgo_hist_unique =  perfil_riesgo_hist_unique.assign(varianza_riesgo_categ = -1)

# Test para asegurarme de que sé encontrar el objeto deseado y modificarlo
perfil_riesgo_hist_unique.loc['33d6a855ebc2a83b45e1aaa480bd169e54b963e0','varianza_riesgo_categ'] = -69
```

```python
# Calculo varianza para riesgo_categ_log para cada individuo
# Si este valor es el mismo en todas las entradas la varianza es 0. 
# Si este valor es único (i.e. solo hay un registro en base long, es decir no se ha modificado el riesgo) la varianza es 0.

for i, row in perfil_riesgo_hist_unique.iterrows():
    # Calculo varianza sobre riesgo_categ_log de base long
    varianza = np.var(perfil_riesgo_hist[perfil_riesgo_hist['afi_hash64']== i ]['riesgo_categ_log'])
    # Guardo varianza calculada en base wide
    perfil_riesgo_hist_unique.loc[i, 'varianza_riesgo_categ'] = varianza
```
